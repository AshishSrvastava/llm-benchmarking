For each of Questions 15 to 19, select the *two* answer choices that, when used to compete the sentence, fit the meaning of the sentence as a whole *and* produce completed sentences that are alike in meaning

```python
import json

def generate_prompt(json_file, question_id):
    with open(json_file) as f:
        data = json.load(f)

    question = data["questions"][question_id]
    prompt = question["prompt"]
    blanks = question["blanks"]
    question_type = question["type"]
    
    if question_type == "text_completion":
        num_blanks = len(blanks)
        prompt_str = f"You are a student taking the GRE solving a text completion question.\n\nThe question is as follows:\n{prompt}\n\nYou need to pick the word best suited for each blank space:"
        
        for blank in blanks:
            prompt_str += f"\n{blank['id']}) "
            for choice in blank["choices"]:
                prompt_str += f"{choice['label']}) {choice['text']}   "
        
        prompt_str += "\n\nChoose a letter from each of the options given that is correct. Those letters are: "
    
    else:
        num_blanks = 0
        prompt_str = f"You are a student taking the GRE.\n\nThe question is as follows:\n{prompt}\n\nYour choices are as follows:"
        
        if num_blanks == 1:
            for blank in blanks:
                prompt_str += f"\n{blank['id']}) "
                for choice in blank["choices"]:
                    prompt_str += f"{choice['label']}) {choice['text']}   "
            prompt_str += "\n\nChoose a letter from the options given that is correct. That letter is: "
        
        elif num_blanks >= 2:
            for blank in blanks:
                prompt_str += f"\n{blank['id']}) "
                for choice in blank["choices"]:
                    prompt_str += f"{choice['label']}) {choice['text']}   "
            prompt_str += "\n\nChoose a letter from each of the options given that is correct. Those letters are: "
        
        else:
            prompt_str = f"Question {question_id} has no blanks. Please choose another question."
    
    return prompt_str
```
